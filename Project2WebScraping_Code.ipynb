```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.edge.options import Options as EdgeOptions
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```python
Title = []
Ratings = []
Reviews  = []
C_Type = []
Salaries = []
Interviews = []
Jobs = []
for i in range(1,24):
    url = f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={i}"
    service = EdgeService(EdgeChromiumDriverManager().install())
    options = EdgeOptions()
    driver = webdriver.Edge(service=service, options=options)
    driver.get(url)
    print(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    company = soup.find_all("div", class_ = "companyCardWrapper")

    for i in company:
        title = i.find("h2", class_ = "companyCardWrapper__companyName")
        titreg = re.findall(r"\s+(.+)\s+",title.text)
        if titreg:
            Title.append(titreg[0])
        else:
            Title.append(np.nan)

        rating = i.find("div", class_ = "rating_star_container")
        ratreg = re.findall(r"\s+(.+)",rating.text)
        if ratreg:
            Ratings.append(ratreg[0])
        else:
            Ratings.append(np.nan)

        reviews = i.find("span", class_ = "companyCardWrapper__companyRatingCount")
        revreg = re.findall(r"\((.+)\)",reviews.text)
        if revreg:
            Reviews.append(revreg[0])
        else:
            Reviews.append(np.nan)

        ctype = i.find("span", class_ = "companyCardWrapper__interLinking")
        typreg = re.findall(r"\s+(.+)\s\|",ctype.text)
        if typreg:
            C_Type.append(typreg[0])
        else:
            C_Type.append(np.nan)

        salary = re.findall(r"Reviews(.+)\sSalaries",i.text)
        if salary:
            Salaries.append(salary[0])
        else:
            Salaries.append(np.nan)
            
        interview = re.findall(r"Salaries(.+)\sInter",i.text)
        if interview:
            Interviews.append(interview[0])
        else:
            Interviews.append(np.nan)

        job = re.findall(r"Interviews(.+)\sJobs",i.text)
        if job:
            Jobs.append(job[0])
        else:
            Jobs.append(np.nan)

        
            
    driver.quit()


# 59.1k Reviews5.9L Salaries8.3k Interviews9.1k Jobs7.4k Benefits39 Photos
```
